```python
#prime
import math
def prime(num):
  if num%2==0 and num>2:
    return False
  for i in range(3,int(math.sqrt(num))+1):
    if num%i==0:
      return False
  return True
prime(7)
```

```python
#LCM

def lcm(x,y):
  if x>y:
    greater=x
  else:
    greater=y
  while True:
    if(greater%x==0) and (greater%y==0):
      lcm=greater
      break
    greater+=1
  return lcm
a=int(input("Enter a number:"))
b=int(input("Enter a number:"))
LCM=lcm(a,b)
print(f"LCM of {a} and {b} = {LCM}")


```

```python
#armstrong number
# 153=1*1*1+5*5*5+3*3*3*3
a=int(input("enter a number:"))
sum=0
temp=a
while temp:
  digit=temp%10
  sum+=digit**3
  temp//=10
if a==sum:
  print("Armstrong number")
else:
  print("no")
```

```python
#calculate power using reccursion
def power(base,expo):
  if expo==1:
    return base
  return base*power(base,expo-1)
base=int(input("enter a  base:"))
expo=int(input("enter a exponent:"))
result=power(base,expo)
print(result)
```
